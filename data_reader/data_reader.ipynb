```python
# パッケージ
import pandas_datareader.data as web
import pandas as pd
import os
from joblib import Parallel, delayed
```

```python
topix_code = pd.read_csv('../data/stock_code/TOPIX_weight_jp.csv',index_col=0)
topix_code = topix_code.sort_values('TOPIXに占める個別銘柄のウェイト',ascending=False)
display(topix_code)

new_dir_path = '../data/raw_data/candle_data/TOPIX'
os.makedirs(new_dir_path, exist_ok=True)


def data_loader(i):
    code = topix_code.iloc[i,topix_code.columns.get_loc('コード')]
    code_name = topix_code.iloc[i,topix_code.columns.get_loc('銘柄名')]
    print(code)
    print(code_name)
    # 株価情報の取得
    df = web.DataReader(f"{int(code)}.T", data_source='yahoo', start='2019-01-01')
    
    # display(df)
    df.to_csv(f'{new_dir_path}/{code_name}.csv')

Parallel(n_jobs=-1)(delayed(data_loader)(i) for i in range(500))
```
